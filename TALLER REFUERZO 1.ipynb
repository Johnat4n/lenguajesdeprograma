**1.** Un vededor recibe un sueldo base mas un 10% por comision de sus ventas , el vendedordesea saber cuanto dienro obtendra por el concepto de comisiones por las tres ventas que realiza en el mes y el total que recibira en el mes tomando en cuenta su sueldo basse y comisiones

```python
#variable para el sueldo base
sueldo_base=float(input(" Digite sueldo base "))
#calcular valor de la comisión
comisión=sueldo_base*0.10
#suma sueldo base mas comisión
total= sueldo_base + comisión
# visualiza sueldo total
print("La comisión es = {:,.0f} y el total del sueldo es = {:,.0f}".format(comisión, total))
print(f"La comisión es = {comisión:,.0f} y el total del sueldo es = {total:,.0f}")
```

**2.** Un maestro desea saber que porcentaje de hombres y que porcentaje de mujeres hay en un grupo de estudiantes

```python
#variable para cantidad de estuiantes de un grupo
cantidad_estudi=float(input(" Digite cantidad de estudiantes"))
#calcular hombres de grupo
porcentaje_H=cantidad_estudi*0.64
#calcular mujeres de grupo
porcentaje_M=cantidad_estudi*0.36
#visualiza cantidad de hombres y mujeres del grupo
print("porcentaje de hombre es = {:,.0f} y procentaje de mujeres es = {:,.0f}".format(porcentaje_H, porcentaje_M))
print(f"porcentaje de hombre es = {porcentaje_H:,.0f} y procentaje de mujeres es = {porcentaje_M:,.0f}")
```
